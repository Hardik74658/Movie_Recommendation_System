###1.3 Loading Database

```python
import pandas as pd

df = pd.read_csv( '//kaggle/input/tmdb-movies-dataset-2023-930k-movies//TMDB_movie_dataset_v11.csv')
df.head()
```

```python
# Keep only 'Released' movies
df = df[df['status'] == 'Released'].copy()
```

```python
# Make sure 'popularity' is numeric
df['popularity'] = pd.to_numeric(df['popularity'], errors='coerce')

# Drop rows with NaN popularity
df_clean = df.dropna(subset=['popularity'])

# Sort by popularity (descending) and take top 20k
top_20k_df = df_clean.sort_values(by='popularity', ascending=False).head(10000)

# Reset index (optional, for clean iteration)
top_20k_df = top_20k_df.reset_index(drop=True)

# Now you can work with this top_20k_df
print(top_20k_df[['title', 'popularity']].head())
```

```python
top_20k_df['title'].isnull().sum()
top_20k_df.dropna(subset=['title'],inplace=True)
```

```python
top_20k_df.info()
```

```python
top_20k_df['overview'] = top_20k_df['overview'].apply(lambda x : str(x).lower())
top_20k_df['genres'] = top_20k_df['genres'].apply(lambda x : str(x).lower())
top_20k_df['spoken_languages'] = top_20k_df['spoken_languages'].apply(lambda x : str(x).lower())
top_20k_df['keywords'] = top_20k_df['keywords'].apply(lambda x : str(x).lower())
top_20k_df
```

```python
top_20k_df['adult']
```

```python
top_20k_df['adult']=top_20k_df['adult'].apply(lambda x: 'adult' if x == True else 'non-adult' )
```

```python
top_20k_df['summary'] = top_20k_df['overview']+" "+top_20k_df['genres']+" "+top_20k_df['keywords']+" "+" "+top_20k_df['spoken_languages']+" "+top_20k_df['adult']
top_20k_df['summary'].isnull().sum()
top_20k_df.drop(columns=['overview','genres','keywords','spoken_languages','adult'])
```

```python
top_20k_df['summary'][0]
```

```python
from sklearn.feature_extraction.text import CountVectorizer

vectorizer = CountVectorizer(stop_words='english',analyzer='word',max_features=5000)
X = vectorizer.fit_transform(top_20k_df['summary'])
```

```python
from sklearn.metrics.pairwise import cosine_similarity
similarity = cosine_similarity(X)
similarity
```

```python
def recommend(movie):
    index = top_20k_df[top_20k_df['title'] == movie].index[0]
    distances = sorted(list(enumerate(similarity[index])),reverse=True,key = lambda x: x[1])
    for i in distances[1:6]:
        print(top_20k_df.iloc[i[0]].title)
```

```python
recommend('The Avengers')
```
